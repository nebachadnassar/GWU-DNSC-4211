## 1.4 Control Flows

Use a for-loop to iterate over a list that you define that computes the modulus of each value in the list divided into 105. Your list should include numbers for 1-10.

```r
for (i in 1:10){
  print(105 %% i)
}
```

```r
divisor <- 105
for (i in 1:10){
  result <- divisor %% i
  print(result)
}
```

Build on the previous question and use an if-statement to only print numbers that divide evenly into 105.

```r
for (i in 1:10){
  mod_result <- 105 %% i
  if (mod_result == 0){
    print(i)
  }
}
```

```r
for (i in 1:10){
  mod_result <- 1053 %% i
  if (mod_result == 1){
    print(i)
  }
}
```

```r
for (i in 1:100){
  mod_result <- 1051 %% i
  if (mod_result == 1){
    print(i)
  }
}
```

```r
divisor <- 105
for (i in 1:10){
  result <- divisor %% i
  if (result == 1){
    print(i)
  }
}
```

```r
divisor <- 105
for (i in 1:10){
  result <- divisor %% i
  if (result == 1){
    print(i)
  }
}
```

```r
mod_func <- function(divisor, l, denominator){
  for (i in l){
    result <- divisor %% i
    if (result == denominator){
      print(i)
  }
}
}
```

```r
mod_func(105, 1:10, 0)
```
