#### DataFrames

Dataframes are generic data objects of R which are used to store the tabular data. Dataframes are the foremost popular data objects in R programming because we are comfortable in seeing the data within the tabular form. They are two-dimensional, heterogeneous data structures. These are lists of vectors of equal lengths.

Data frames have the following constraints placed upon them:

- A data-frame must have column names and every row should have a unique name.
- Each column must have the identical number of items.
- Each item in a single column must be of the same data type.
- Different columns may have different data types.


````r
#%%R
#-- This is probably the most important native Data
#-- Structure to R users.
#-- This is also a big difference between R and Python
#-- because Python does not have a native DataFrame Structure

Countries     = c("Cayman Islands", "BVI", "Cyprus", "Liberia")
Region        = c("Caribbean", "Caribbean", "Europe", "Africa")
Bank_Assets   = c(1000000, 500000, 750000, 10000)
GDP           = c(100, 80, 70, 50)

df = data.frame(Countries, Region, Bank_Assets, GDP)

print(df)
````

````r
mean(df$Bank_Assets)
````

## 1.4 Control Flows: If-Statements and Loops

Resource: https://adv-r.hadley.nz/control-flow.html

In R, as in pretty much every programming language, there exist if-statements and various types of loops to control the flow of code execution. You will notice, that in contrast to Python you will be using curcly brackets instead of semi-colons and spacing to execute code.

R If-Statements and Loop Types
- If-Else
```
if (expression) {
    # execute these statements
} else {
    # execute these statements
}
Note: The else must be in the same line as the closing braces of the if statement.
```
- For-Loop
```
for iterable {
    # execute these statements
}
```
- While-Loop
```
while (expression){
    statement(s)
}
```


#### If-Statement

````r
#-- structure of the R if-statement is very similar to Python
#----- core components
#- 1. if key word
#- 2. some boolean comparison inside parenthesis
#- 3. opening and closing curly braces that define the
#     body of the if-statement
#- 4.  Something in the body!

if (TRUE) {
  print("Code executed!")
}
````

````r
if (TRUE) { print("Code executed!") }
````

````r
#-- Remember R doesnt depend on spacing like Python does
#-- You could, in theory, write an entire if statement in a
#-- single line

#-- but, why would you ever want such poor readability
a <- 10
if (a >= 11){
  print("Code executed!")
}
````

````r
#-- in R (as in Python) you can have any variation of if, if-else, or
#-- if, else if, else statements
letter_grade = NULL
numeric_grade <- 79
if (numeric_grade >= 90) {
  letter_grade <- "A"
} else if (numeric_grade >= 80){
  letter_grade <- "B"
} else {
  letter_grade <- "Not A or B"
}
print(letter_grade)
````

````r
#-- Remember! 2 if statements are different than and if-else statement
numeric_grade <- 90
if (numeric_grade >= 90) {
  letter_grade <- "A"
}
if (numeric_grade >= 80) {
  letter_grade <- "B"
}
print(letter_grade)
````

````r
numeric_grade <- 90
if (numeric_grade >= 90) {
  letter_grade <- "A"
}
if (numeric_grade >= 80 & numeric_grade < 90) {
  letter_grade <- "B"
}
print(letter_grade)
````

````r
#%%R
#-- You can have an if statement, an if-else statement, and an if-else if-else
#-- statement/

#-- In the example below, also note that R does automatic conversion when
#-- comparing data

b <- 11
a <- 10

if (b > a) {
  print("b is greater than a")
} else if (a == b) {
  print("a and b are equal")
} else {
  print("a is greater than b")
}
````

````r
#%%R
# In coding, you will always have many ways
# to complete the same task. There often exist decisions
# that have trade-offs in simplicity and efficiency.
x <- 91
if (x > 90) {
  grade <- "A"
}
if (x > 80 & x <= 90) {
  grade <- "B"
}

if (x > 70 & x <= 80) {
  grade <- "C"
}
if (x <= 70) {
  grade <- "F"
}
print(grade)
````
