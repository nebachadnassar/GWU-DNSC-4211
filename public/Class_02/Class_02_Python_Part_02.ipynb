### The Basics

#### Syntax for Functions

You use the keyword def to tell Python that you are ready to define a function, followed by the name that you want to name the function. In this case, we name it "add". Following the name of the function you should have opening and closing parenthesis. Optionally you can allow the user to input data by adding variables into the parenthesis, that may be needed in the function. These user-passed variables are called **arguments** or **parameters**. The terms are interchangeable. We will see some examples later without such arguments
```python
# Syntax for a basic function to add two numbers
def add(x, y):
  #-- Code that you want to execute in the function
  s = x + y

  #-- then we need to return the result back to the user
  return s
```

Now that we have a function the syntax to call a function is
```python
summation = add(2, 4)
print(summation)
```
```ipython
output [0]: 6
```
Alternatively, you can also have optional/default arguments that the user of
the function can choose to use or not. The below example defaults to adding 10 to x. You do this by setting the parameter equal to a default value. When you have a mix of mandatory and default arguments, the default arguments **MUST GO AFTER THE MANDATORY ARGUMENTS**
```python
#-- Syntax for a basic function to add two numbers
def add(x, y=10):
  #-- Code that you want to execute in the function
  s = x + y

  #-- then we need to return the result back to the user
  return s
print(add(18))
```
```ipython
output [0]: 28
```

### Testing functions out with prime numbers

````python
def is_prime(input_num=97):
  is_div = 0
  for i in range(2, input_num):
    if input_num % i == 0:
      return False # Return make the program stop running
  return True
````

````python
prime = is_prime(input_num=101)
prime
````

````python
#-- How to turn this code into a function that we can use repeatably?
prime_numbers = []
for i in range(3, 101):
  tests = []
  for j in range(2, i):
    tests.append(i % j == 0)
  if sum(tests) == 0:
    prime_numbers.append(i)
print(prime_numbers)
````

````python
# This will do... but what if we want to change the range for which we search?

# We can define a function wtih arguments

# Comments within the function are great, as well as detailed descriptions
# of what they are

def find_prime_numbers(start_num, end_num):
    '''
      @start_num: Starting integer to test if it is prime
      @end_num:   Ending integer (inclusive) to test if it is prime
    '''
    prime_numbers = []
    #-- Add 1 because to include it in the check
    for i in range(start_num, end_num +1):
      tests = []
      #-- Iterate from 3 up to current number that is i
      for j in range(2, i):
        #-- Append boolean value to list True, if i is divisible by a number
        tests.append(i % j == 0)
      #-- Test if all values are False e.g. prime
      if sum(tests) == 0:
        prime_numbers.append(i)
    return prime_numbers

#-- Call the function
prime_numbers = find_prime_numbers(3, 148)
print(prime_numbers)
````

### Functions with optional parameters

````python

# Additionally, when a function runs a return statement,
# the function will immediately stop running
def names_to_string(first_name, last_name, middle_name=""):
  '''
    This function takes as input a first, last and middle name and prints
    them together in one string

    @first_name: mandatory String argument representing a first name
    @last_name: mandatory String argument representing a last name
    @middle_name: optional String argument representing a middle name
  '''
  print("{} {} {}".format(first_name, middle_name, last_name))
  return None #-- This is optional

names_to_string("Bob", "Ross") #-- Example not using the optional argument
names_to_string("Bob", "Ross", "Jay") #-- Example using the optional argument
names_to_string(first_name="Bob", last_name="Ross", middle_name="Jay") #-- Example with explicitly stating arguments

````

## 3.3 Advanced Control Flows: Exception Handling, continue and pass

A well-written program has the ability to work no matter what. A key component of this is the ability to direct the program what to do when handling errors. In Python, you can use try-excep statements that work very similarly to if-else statements.

```
try:
    # Some Code
except:
    # Executed if error in the
    # try block
```

### Try-Except Statement

````python
# For Question #2 on ASSN_1, many of you probably had to
# use string methods to try and figure out if a variable
# is able to be converted to a numeric value or not.
# an alternative method, would be to just try and convert
# every value and see if it throws an error.
list = ["1", "1.3", "1B", "4", "tent"]
s = 0
for i in list:
  try:
    s += float(i)
  except:
    print('ERROR', i)
print(s)
````

````python
# You can specify handling for any error type or, you
# can specify specific errors
list = ["1", "1.3", "1B", "4", "tent"]
s = 0
for i in list:
  try:
    s += float(i)
  except ValueError:
    pass
print(s)
````

### Continue, pass, break
- Break:A break statement in Python alters the flow of a loop by terminating it once a specified condition is met.
- Continue: The continue statement in Python is used to skip the remaining code inside a loop for the current iteration only.
- Pass: The pass statement in Python is used when a statement or a condition is required to be present in the program, but we dont want any command or code to execute. Its typically used as a placeholder for future code.

````python
# when the break statement is used, the infinite loop is broken
# break will only stop the inner loop however
i = 0
while True:
  if i >= 5:
    break
  print(i)
  i += 1
````

````python
# The break will only break it's direct parent
i = 0
for _ in range(5):
  while True:
    if i >= 5:
      break
    print("spam")
    i += 1
  print("more spam")
````

````python
# continue skips only the iteration it is on, but will continue
# Notice 5 is not printed
for num in range(5):
    if num == 3:
        continue
    print(f'Iteration: {num}')
````

````python
# pass does nothing, it is useful for development because you can have an if
# statement and not have any code in it
if 1 > 5:
  pass
else:
  pass
````
